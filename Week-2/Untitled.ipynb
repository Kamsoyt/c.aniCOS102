```python
#cubic equation,quartic equation,quadratic equation calculator

print("Choose equation type")
print("1. Cubic equation(ax^3 + bx^2 + cx + d = 0)")
print("2. quartic equation(ax^4 + bx^3 + cx^2 + dx + c = 0)")
print("3. quadratic equation(ax^2 + bx + c = 0)")

choice = int(input("Enter choice: "))

#cubic
if choice == 1:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))

    print("Approximate real root:")

    for x in range(-50, 51):
        y = a*x**3 + b*x**2 + c*x + d
        if abs(y) < 1:
            print("possible root near:", x)

#Quartic
elif choice == 2:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))
    e = float(input("Enetr e: "))

    print("Approximate real root:")

    for x in range(-50, 51):
        y = a*x**4 + b*x**3 + c*x**2 + d*x + e 
        if abs(y) < 1:
            print("possible root near:", x)

#Quadratic
elif choice == 3:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))

    print("Approximate real root:")

    for x in range(-50, 51):
        y = a*x**2 + b*x + c 
        if abs(y) < 1:
            print("possible root near:", x)


else:
    print("Invalid choice")
```
